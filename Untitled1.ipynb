```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [50,85,120,45,95,70, 110,65,130,100,90],
    [2,3,4,1,3,2,4,2,5,4,3],
    [5,8,12,4,10,6,15,7,20,9,14]
]).T
y = np.array([[75000,120000,200000,60000,140000,100000,180000,90000,230000,160000,150000]]).T
```

```python
X_1 = X[:,0].reshape(-1,1)
X_2 = X[:,1].reshape(-1,1)
X_3 = X[:,-1].reshape(-1,1)
```

```python
one = np.ones((X.shape[0],1))
X_bar_1 = np.concatenate((one,X_1), axis = 1)
X_bar_2 = np.concatenate((one,X_2), axis = 1)
X_bar_3 = np.concatenate((one,X_3), axis = 1)
```

```python
A_1 = np.dot(X_bar_1.T,X_bar_1)
A_2 = np.dot(X_bar_2.T,X_bar_2)
A_3 = np.dot(X_bar_3.T,X_bar_3)
B_1 = np.dot(X_bar_1.T, y)
B_2 = np.dot(X_bar_2.T, y)
b_3 = np.dot(X_bar_3.T, y)
```

```python
w_1 = np.dot(np.linalg.pinv(A_1), B_1)
w_2 = np.dot(np.linalg.pinv(A_2), B_2)
w_3 = np.dot(np.linalg.pinv(A_3), b_3)
```

```python
X_0_1 = np.linspace(50, 130, 2)
X_0_2 = np.linspace(1, 5, 2)
X_0_3 = np.linspace(4, 20, 2)
y_0_1 = w_1[0][0] + w_1[1][0]*X_0_1
y_0_2 = w_2[0][0] + w_2[1][0]*X_0_2
y_0_3 = w_3[0][0] + w_3[1][0]*X_0_3
```

```python
feature_list = ['Area(m^2)', 'Room Count', 'distance to the center' ]
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,4))
for i in range(3):
  ax[i].plot(eval('X_'+str(i+1)), y, 'ro', )
  ax[i].plot(eval('X_0_'+str(i+1)), eval('y_0_'+str(i+1)))
  ax[i].set_xlabel(f'{feature_list[i]}')
  ax[i].set_ylabel('House_price')
  ax[i].set_title(f'Độ tương qua giữa House_price và {feature_list[i]}')
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, wspace=0.4, hspace=0.3)
plt.show()
```
